```python
def generate_youtube(channelID, num_vid=20):
    '''Function to get the Youtube videos and seperate them to DataFrame
    
    Parameters:
        channelID (str): The id of the Youtube channel.
        num_vid (int): Number of videos to check. Default is 20.

    Returns:
        youtube (pandas.DataFrame): The last videos' details.
    '''
    
    YoutubeAPIkey = "AIzaSyBNUKHYrWPZkNjJoEfOyhUFXM0VuZnEO7U"
    youtube_url = "https://www.googleapis.com/youtube/v3/search?key=" + \
                    YoutubeAPIkey + "&channelId=" + \
                    channelID + "&part=snippet,id&order=date&maxResults=" + str(num_vid)
    
    # Get the URL for info
    page = requests.get(youtube_url)
    
    # get the items from googleapis
    result = json.loads(page.text)
    google = pd.json_normalize(result["items"])
    
    # Filter necessary info from googleapis
    youtube = google.filter(['id.videoId','snippet.publishedAt',\
                             'snippet.channelId', 'snippet.title'], axis=1)
    
    # Create and add the video links
    declar = "http://www.youtube.com/watch?v="
    links = declar + youtube["id.videoId"]
    youtube['link'] = links
    youtube.rename(columns = {'id.videoId': '_id',\
                              'snippet.publishedAt': 'date',\
                              'snippet.channelId': 'channelId',\
                              'snippet.title': 'title' ,\
                              'snippet.comment': 'comment'}, inplace=True)
    
    return youtube
```
